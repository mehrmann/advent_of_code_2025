```kotlin
import java.nio.file.Files
import java.nio.file.Paths

val input = Files.readString(Paths.get("input.txt"))
```

```kotlin
val parts = input.trim().split("\n\n")
val valid = parts[0].splitToSequence("\n").map {
    val ids = it.split("-").map { it.toULong()}
    ids[0]..ids[1]
}.toList()

val ingredients = parts[1].splitToSequence("\n").map {
    it.toULong()
}

ingredients.fold(0) { acc, ingredient ->
    val notSpoiled = valid.firstOrNull { ingredient in it }
    acc + if (notSpoiled!=null) 1 else 0
}
```

```kotlin
val allIngredients = mutableListOf<ULongRange>()
val sortedValid = valid.sortedBy { it.first }
var current = sortedValid[0]
sortedValid.drop(1).forEach { next ->
    if (next.first < current.last + 1UL) {
        current = current.first..maxOf(current.last, next.last)
    } else {
        allIngredients.add(current)
        current = next
    }
}
allIngredients.add(current)

val num = allIngredients.fold(0UL) { acc, range ->
    acc + range.last - range.first + 1UL
}

num
```
