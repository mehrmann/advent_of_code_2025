```kotlin
val test_input =
"123 328  51 64 \n" +
" 45 64  387 23 \n" +
"  6 98  215 314\n" +
"*   +   *   +  \n"
```

```kotlin
import java.nio.file.Files
import java.nio.file.Paths

val use_input = true
val seq = if (use_input) {
    Files.readString(Paths.get("input.txt"))
} else {
    test_input
}.split('\n').toList().dropLast(1)
```

# Part 1

```kotlin
val spaceRegex = Regex("\\s+")

val n = seq.dropLast(1).map {
    it.trim().split(spaceRegex).map { it.toULong() }
}
val ops = seq.last().split(spaceRegex)

//transpose
n[0].indices.map { col -> n.map { it[col]} }
    //sum
    .foldIndexed(0UL) { index, acc, longs ->
        //apply operations to list of longs
        acc + longs.drop(1).fold(longs[0]) { acc, lng ->
            if (ops[index] == "*") {
                acc * lng
            } else {
                acc + lng
            }
        }
    }
```

# Part 2

```kotlin
//find starting positions in list of operators
val positions = seq.last().flatMapIndexed { i,s -> if (s != ' ') listOf(i) else emptyList()}
//find distance to next operator
val widths = positions.zipWithNext { a,b -> b-a - 1} + (seq.last().length - positions.last())
//calculate ranges
val ranges = positions.zip(widths) { pos,len -> pos..(pos+len-1) }

//dissect number data into substrings based on ranges
val cols = ranges.map { range ->
    seq.dropLast(1).map { it.substring(range) }
}

//transpose the number data and create top down numbers
val transposed = cols.map { l ->
    l[0].indices.map { i -> l.map { it[i]}.joinToString("").trim().toULong() }
}

//apply the operations and create a sum
transposed.foldIndexed(0UL) { index, acc, longs ->
    acc + longs.drop(1).fold(longs[0]) { acc, lng ->
        if (ops[index] == "*") {
            acc * lng
        } else {
            acc + lng
        }
    }
}
```
