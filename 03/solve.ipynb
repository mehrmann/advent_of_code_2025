```kotlin
fun findMaxJoltage(seq: List<Int>) : Int {

    val firstDigit = seq.subList(0, seq.size-1).withIndex().maxBy { it.value }
    val secondDigit = seq.drop(firstDigit.index+1).withIndex().maxBy { it.value }
    return firstDigit.value * 10 + secondDigit.value
}

val testData = mapOf(
    "987654321111111" to 98,
    "811111111111119" to 89,
    "234234234234278" to 78,
    "818181911112111" to 92
)

testData.forEach { i, o ->
    val seq = i.toList().map { Integer.parseInt(it.toString())}
    val res = findMaxJoltage(seq)
    println("$i -> $res ($o)")
    check(res == o)
}

```

```kotlin
import java.nio.file.Files
import java.nio.file.Paths

Files.readString(Paths.get("input.txt"))
    .split('\n')
    .filter { it.isNotEmpty() }
    .asSequence()
    .mapNotNull {
        it.trim().toList().mapNotNull { Integer.parseInt(it.toString())}
    }.fold(0) { acc, seq ->
        //println("${seq.joinToString("")} -> ${findMaxJoltage(seq)}")
        acc + findMaxJoltage(seq)
    }
```

```kotlin
fun findMaxJoltage(seq: List<Int>, num: Int) : ULong {
    var packs = mutableListOf<Int>()
    ((num-1).downTo(0)).fold(0) { start, n ->
        val leftToPickFrom = seq.size - start - n

        val toSearch = seq.subList(start, start + leftToPickFrom)
        //println("considering left=$leftToPickFrom $start, $n -> ${toSearch}")
        val res = toSearch.withIndex().maxBy { it.value }
        //println("picked ${res.value} at ${res.index}")
        packs.add(res.value)
        res.index+1 + start
    }
    return packs.joinToString("").toULong()
}

val testData = mapOf(
    "987654321111111" to 987654321111UL,
    "811111111111119" to 811111111119UL,
    "234234234234278" to 434234234278UL,
    "818181911112111" to 888911112111UL
)

testData.forEach { i, o ->
    val seq = i.toList().map { Integer.parseInt(it.toString())}
    val res = findMaxJoltage(seq, 12)
    println("$i -> $res ($o)")
    check(res == o)
}
```
