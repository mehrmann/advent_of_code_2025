```kotlin
import java.nio.file.Files
import java.nio.file.Paths

val input = Files.readString(Paths.get("input.txt"))
    .split(',')
    .asSequence()
    //map to ranges
    .map {
        val elements = it.trim().split('-')
        elements[0].toULong()..elements[1].toULong()
    }
```

```kotlin

    //count invalid ids
input.fold(0UL) { invalid, seq ->
    invalid + seq.fold(0UL) { acc, i ->
        val s = i.toString()
        //an id is invalid when the front half and the back half are identical
        val a = s.substring(0, s.length / 2)
        val b = s.substring(s.length / 2)
        if (a == b) {
            acc + i
        } else {
            acc
        }
    }
}
```

```kotlin
input.fold(0UL) { invalid, seq ->
    invalid + seq.fold(0UL) { acc, i ->
        val s = i.toString()
        var invalid = false
        //now also check repetitions of subsequences
        for (split in 1..<s.length) {
            val chunked = s.chunked(split)
            //how many are the same like the first chunk?
            val count = chunked.count { chunked[0] == it }
            //if the count is the number of chunks they all are the same
            if (count == chunked.size) {
                invalid = true
                //println("$s is invalid (${chunked[0]} * ${chunked.size})")
                break
            }
        }
        if (invalid) {
            acc + i
        } else {
            acc
        }
    }
}
```

```kotlin
input.fold(0UL) { invalid, seq ->
    invalid + seq.fold(0UL) { acc, i ->
        val s = i.toString()
        var invalid = false
        for (split in 1..s.length/2) {
            // Only check splits that evenly divide the length
            if (s.length % split == 0) {
                val chunked = s.chunked(split)
                val first = chunked[0]
                //this breaks earlier when first "mismatch" is detected
                if (chunked.all { it == first }) {
                    invalid = true
                    break
                }
            }
            if (invalid) break
        }
        if (invalid) acc + i else acc
    }
}
```

```kotlin
//Alternative more functional approach.
//gets rid of intermediate state variables
//"Is there ANY split where ALL positions match?"

input.fold(0UL) { invalid, seq ->
    invalid + seq.fold(0UL) { acc, i ->
        val s = i.toString()

        val invalid = (1..s.length / 2).any { split ->
            s.length % split == 0 &&
                    (split until s.length step split).all { pos ->
                        s.regionMatches(0, s, pos, split)
                    }
        }

        if (invalid) acc + i else acc
    }
}
```
