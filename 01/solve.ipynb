```kotlin
var zeros = 0
Files.readString(Paths.get("input.txt"))
    .split('\n')
    .asSequence()
    .mapNotNull {
        val s = it.trim()
        if (s.isNotEmpty()) {
            val sign = if (s[0] == 'L') -1 else 1
            sign * s.substring(1).toInt()
        } else null
    }.fold(50) { dial, turn ->
        val new = (dial + turn) % 100
        if (new == 0) {zeros +=1}
        new
    }
zeros
```

```kotlin
var zeros = 0
Files.readString(Paths.get("input.txt"))
    .split('\n')
    .asSequence()
    .mapNotNull {
        val s = it.trim()
        if (s.isNotEmpty()) {
            val sign = if (s[0] == 'L') -1 else 1
            sign * s.substring(1).toInt()
        } else null
    }.fold(50) { dial, turn ->
        var rest = turn
        var newDial = dial
        while (rest.sign != 0) {
            newDial += rest.sign
            newDial = newDial % 100
            if (newDial == 0) {
                zeros += 1
            }
            rest -= rest.sign
        }
        newDial
    }
zeros
```
